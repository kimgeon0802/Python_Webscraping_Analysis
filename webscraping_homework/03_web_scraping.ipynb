#### 1. 질문 :  검색어로  찾은  책 목록을 json 파일로 저장하기
   data/books.json 파일로 저장해 주세요.

```python
import requests
from pprint import pprint
import json
from dotenv import load_dotenv
import os

def search_books(query):
    CLIENT_ID = os.getenv("CLIENT_ID")
    CLIENT_SECRET = os.getenv("CLIENT_SECRET")
    headers = {
    'X-Naver-Client-Id': CLIENT_ID,
    'X-Naver-Client-Secret': CLIENT_SECRET,
    }   

    payload = {
        'query':  query, #'파이썬',
        'display': 100,
        'sort': 'sim'
        }

    url = 'https://openapi.naver.com/v1/search/book.json'
    res = requests.get(url, params=payload, headers=headers)
    items_data = res.json()['items']

    #pprint(res.json()['items'])
    #data폴더 경로가 이미 상위에 존재, 상대경로로 지정했습니다.
    with open('../data/books.json','w', encoding='utf-8') as file:
        json.dump(items_data, file)

search_books("파이썬")
```

#### 2. books.json 파일을 Pandas DataFrame로 저장하기

```python
import pandas as pd

data = pd.read_json('../data/books.json')
```

#### 4. 질문 :  검색어로  찾은  책 목록 중에서 가격이 2만원 이상인 책만 출력하기
title,author,discount,publisher,pubdate 컬럼만 출력
가격은 descending (내림차순), index 초기화

```python
data.loc[data['discount'] >= 20000,['title','author','discount','publisher','pubdate']].sort_values(by='discount',ascending=False).reset_index(drop=True)
```
