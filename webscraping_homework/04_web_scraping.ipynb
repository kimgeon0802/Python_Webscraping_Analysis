1. 지역별 총인구수 Top5


```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

data = pd.read_csv('../data/인구현황.csv')

data.loc[data['행정기관']!="전국",['행정기관','총인구수']].sort_values(by='총인구수',ascending=False).head()
```

2. 세대당 인구수 높은 지역 Top5

```python
fig_data1 = data.loc[data['행정기관']!="전국",['행정기관','세대당 인구']].sort_values(by='세대당 인구',ascending=False).reset_index(drop=True)
fig_data1.head()
```

2-1. 지역별 세대당 인구 Plot

```python
figure, axes1 = plt.subplots(nrows=1,ncols=1)
figure.set_size_inches(18,6)

font_path = 'C:\\Windows\\Fonts\\malgun.ttf'
font_name=fm.FontProperties(fname=font_path).get_name()

pop_plot1 = sns.barplot(data=fig_data1.sort_values(by="세대당 인구",ascending=False), x="행정기관",y="세대당 인구",ax=axes1, hue="행정기관")
for item in axes1.get_xticklabels(): 
    item.set_rotation(45)

plt.rcParams['font.family'] =font_name
plt.rcParams['axes.unicode_minus'] =False
pop_plot1.set_title(f'지역별 세대당 인구')
plt.show()
```

3. 남녀 비율 분석

```python
fig_data2 = data.loc[data['행정기관']!="전국",['행정기관','남여 비율']].sort_values(by='남여 비율',ascending=False).reset_index(drop=True)
fig_data2.head(17)
```

3-1. 남녀 비율 분석 Plot
 기준이 되는 빨간선은 pyplot의 axhline() 함수를 사용해 보세요. ( 선택 )

```python
figure, axes1 = plt.subplots(nrows=1,ncols=1)
figure.set_size_inches(18,6)


pop_plot2 = sns.barplot(data=fig_data2.sort_values(by="남여 비율",ascending=False), x="행정기관",y="남여 비율",ax=axes1, hue="행정기관")
plt.axhline(y=1, color='r', linestyle='--', linewidth=2, label='남여 1:1 비율')
for item in axes1.get_xticklabels(): 
    item.set_rotation(45)
    
pop_plot2.set_title(f'지역별 남여 비율')
plt.xlabel('')
plt.legend()
plt.show()
```

4. 남자초과 / 여자초과 분석

```python
fig_data3 = data.loc[data['행정기관']!="전국",['행정기관','남여 비율']].reset_index(drop=True)
fig_data3.index = [x for x in range(1,18)]

fig_data3['남초여초'] = fig_data3['남여 비율'].map(lambda x: "남초" if x > 1 else ("여초" if x < 1 else "동일"))
fig_data3.head(17)
```

5. 세대당 인구 평균보다 높은 지역

```python
a = data.loc[data['세대당 인구'] > data['세대당 인구'].mean() ,["행정기관","세대당 인구"]].sort_values(by='세대당 인구',ascending=False).reset_index(drop=True)
a.loc[a['행정기관'] != "전국", ["행정기관","세대당 인구"]]
```

5-1.   남초 vs 여초 개수 Plot  ( seaborn의 countplot )
         지역별 세대수 Plot ( seaborn의 barplot )


```python
figure, (axes1, axes2) = plt.subplots(nrows=2,ncols=1)
figure.set_size_inches(18,12)

fig_data4 = data.loc[data['행정기관']!="전국",["행정기관","세대수"]]
seaborn_plot = sns.countplot(data=fig_data3,x="남초여초",ax=axes1,hue="남초여초")
barp_plot = sns.barplot(data=fig_data4.sort_values(by="세대수",ascending=False), x="행정기관",y="세대수",ax=axes2, hue="행정기관")

seaborn_plot.set_title(f'남초 vs 여초 지역 개수')
barp_plot.set_title(f'지역별 세대수')
plt.show()
```

6. 남자 인구 vs 여자 인구 비교 Plot ( 선택 )

```python
import numpy as np

figure, axes1 = plt.subplots(nrows=1,ncols=1)
figure.set_size_inches(18,12)
bar_width = 0.25
m_data = data.loc[data['행정기관']!="전국",["행정기관","남자 인구수","여자 인구수"]]
N = m_data.shape[0]
index = np.arange(N)

m = plt.bar(index-bar_width, m_data['남자 인구수'], bar_width, alpha=0.4, color='blue', label='남자 인구수')
w = plt.bar(index, m_data['여자 인구수'], bar_width, alpha=0.4, color='red', label='여자 인구수')

plt.xticks(index,m_data["행정기관"])
plt.title(f'남자 vs 여자 인구 비교')
plt.legend()
plt.show()
```
